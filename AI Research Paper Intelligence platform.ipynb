20. Creating Search Function

```python
from re import S
def search_paper(query, k = 5):
  query_embedding = model.encode([query])
  faiss.normalize_L2(query_embedding)
  D,I = index.search(query_embedding, k)
  for score, idx in zip(D[0],I[0]):
    print("Similarity Score", score)
    print("Title", df.iloc[idx]["title"])
    print("Title", df.iloc[idx]["abstract"][:500])
    print()
```

APPROACH 2: Named Entity Recognition using Zero-Shot Classification and Transformers

```python
!pip install spacy transformers torch
!python -m spacy download en_core_web_sm
```

```python
import spacy
from transformers import pipeline
```

```python
nlp = spacy.load("en_core_web_sm")
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)
```

```python
candidate_labels = [
    "PROGRAMMING_LANGUAGE",
    "FRAMEWORK",
    "LIBRARY",
    "MODEL",
    "METHOD",
    "APPLICATION",
    "TASK",
    "VECTOR_DATABASE"
]
```

```python
def extract_entities(text):
    doc = nlp(text)
    entities = set()
    for ent in doc.ents:
        entities.add(ent.text)
    for chunk in doc.noun_chunks:
        entities.add(chunk.text)
    for entity in entities:
        result = classifier(
            entity,
            candidate_labels
        )
        label = result["labels"][0]
        print(entity, "->", label)
```

```python
text = """The Semantic Research Paper Search Engine is developed using Python and Sentence Transformers.
FAISS is used to perform semantic similarity search on research papers.
The abstracts are summarized using the BART model and important keywords are extracted using KeyBERT.
Deep learning techniques are applied for medical image analysis and natural language processing tasks."""
```

```python
extract_entities(text)
```
